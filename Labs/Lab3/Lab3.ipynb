```python
# Some Notes about the Isochrone Data
# DATA From   
# http://stellar.dartmouth.edu/models/isolf_new.html
# files have been modified from download.  
# ( M/Mo --> M;   Log L/Lo --> L)
# removed #'s from all lines except column heading
# NOTE SETTINGS USED:  
# Y = 0.245 default   [Fe/H] = -2.0  alpha/Fe = -0.2
# These could all be changed and it would generate 
# a different isochrone
```

```python

# Filename for data with Isochrone fit for 1 Gyr
# These files are located in the folder IsochroneData
filename1="./IsochroneData/Isochrone1.txt"
# Filename for data with Isochrone fit for 10 Gyr
filename10="./IsochroneData/Isochrone10.txt"
# Filename for data with Isochrone fit for 11 Gyr
filename11="./IsochroneData/Isochrone11.txt"
# Filename for data with Isochrone fit for 7 Gyr
filename7="./IsochroneData/Isochrone7.txt"
# filename for data with Isochrone fit for 6 Gyr
filename6="./IsochroneData/Isochrone6.txt"
```

```python
# READ IN DATA
# "dtype=None" means line is split using white spaces
# "skip_header=8"  skipping the first 8 lines 
# the flag "names=True" creates arrays to store the date
#       with the column headers given in line 8 

# Read in data for an isochrone corresponding to 1 Gyr
data1 = np.genfromtxt(filename1,dtype=None,
                      names=True,skip_header=8)
# Read in data for an isochrone corresponding to 10 Gyr
data10 = np.genfromtxt(filename10,dtype=None,names=True,skip_header=8)
# Read in data for an isochrone corresponding to 8 Gyr
data7 = np.genfromtxt(filename7,dtype=None,names=True,skip_header=8)
# Read in data for an isochrone corresponding to 11 Gyr
data11 = np.genfromtxt(filename11,dtype=None,names=True,skip_header=8)
# Read in data for an isochrone corresponding to 6 Gyr
data6 = np.genfromtxt(filename6,dtype=None,names=True,skip_header=8)

```

```python
# Plot Isochrones 
# For Carina

fig = plt.figure(figsize=(10,10))
ax = plt.subplot(111)

# Plot Isochrones

# Isochrone for 1 Gyr
# Plotting Color vs. Difference in Color 
plt.plot(data1['B']-data1['R'], data1['R'], color='blue', 
         linewidth=5, label='1 Gyr')
###EDIT Here, following the same format as the line above 
plt.plot(data7['B']-data7['R'], data7['R'], color='red', 
         linewidth=5, label='7 Gyr')

plt.plot(data10['B']-data10['R'], data10['R'], color='green', 
         linewidth=5, label='10 Gyr')

plt.plot(data11['B']-data11['R'], data11['R'], color='orange',linewidth=5, label='11 Gyr')
plt.plot(data6['B']-data6['R'], data6['R'], color='purple',linewidth=5, label='6 Gyr')

# Add axis labels
plt.xlabel('B-R', fontsize=22)
plt.ylabel('M$_R$', fontsize=22)

#set axis limits
plt.xlim(-0.5,2)
plt.ylim(5,-2.5)

#adjust tick label font size
label_size = 22
matplotlib.rcParams['xtick.labelsize'] = label_size 
matplotlib.rcParams['ytick.labelsize'] = label_size

# add a legend with some customizations.
legend = ax.legend(loc='upper left',fontsize='x-large')

#add figure text
plt.figtext(0.5, 0.15, 'CMD for Carina dSph', fontsize=22)

plt.savefig('/Users/swapnaneeldey/Desktop/ASTR400Bfiles/ASTR400B/Labs/Lab3/IsochroneCarina.png')     
```
